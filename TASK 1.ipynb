## CONTACTMASTER

```python
contacts = []

# Function to add a contact
def add_contact():
    name = input("Enter contact name: ")
    phone = input("Enter contact phone number: ")
    contacts.append({"name": name, "phone": phone})
    print(f"Contact '{name}' added with phone number '{phone}'.")
```

```python
# Function to delete a contact
def delete_contact():
    name = input("Enter contact name to delete: ")
    for contact in contacts:
        if contact['name'] == name:
            contacts.remove(contact)
            print(f"Contact '{name}' deleted.")
            return
    print(f"Contact '{name}' not found.")
```

```python
# Function to search for a contact
def search_contact():
    name = input("Enter contact name to search: ")
    for contact in contacts:
        if contact['name'] == name:
            print(f"Contact '{name}' found with phone number: {contact['phone']}.")
            return
    print(f"Contact '{name}' not found.")
```

```python
# Function to view all contacts
def view_contacts():
    if contacts:
        print("\n--- Contact List ---")
        for contact in contacts:
            print(f"Name: {contact['name']}, Phone: {contact['phone']}")
    else:
        print("No contacts available.")
```

```python
# Main loop to run the application
def main():
    while True:
        print("\n Add Contact\n Delete Contact\n Search Contact\n View Contacts\n5 Exit")
        choice = input("Choose an option : ")
        
        if choice == 'Add Contact':
            add_contact()
        elif choice == 'Delete Contact':
            delete_contact()
        elif choice == 'Search Contact':
            search_contact()
        elif choice == 'View Contacts':
            view_contacts()
        elif choice == 'Exit':
            print("Exiting ContactMaster.")
            break
        else:
            print("Invalid choice. Try again.")

# Run the application
if __name__ == "__main__":
    main()
```
